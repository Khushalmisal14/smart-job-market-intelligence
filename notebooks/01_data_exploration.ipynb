```python
# import the libraries 
import pandas as pd
import numpy as np

# loading the datasets
df = pd.read_csv('../data/raw/ds_salaries.csv')

# first 5 rows
df.head()
```

```python
# Basics info of datasets
df.info()
```

```python
# Checking the Missing Values
df.isnull().sum()
```

```python
# Analysis the Statistical Summary of Dataset
df.describe()
```

```python
# How many values are available in column
print("Job Titles:", df['job_title'].nunique())
print("Countries:", df['company_location'].nunique())
print("Experience Levels:", df['experience_level'].unique())
print("Company Sizes:", df['company_size'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count Top 10 Job Titles 
top_jobs = df['job_title'].value_counts().head(10)

# Draw Graph 
plt.figure(figsize=(12, 6))
sns.barplot(x=top_jobs.values, y=top_jobs.index, palette='viridis')
plt.title('Top 10 Most Common Job Titles')
plt.xlabel('Number of Jobs')
plt.ylabel('Job Title')
plt.tight_layout()
plt.show()
```

```python
# Salary Distribution Graph
plt.figure(figsize=(12, 6))
sns.histplot(df['salary_in_usd'], bins=50, color='steelblue', kde=True)
plt.title('Salary Distribution of Data Science Jobs')
plt.xlabel('Salary in USD')
plt.ylabel('Number of Jobs')
plt.tight_layout()
plt.show()
```

```python
# Experience Level vs Salary
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', 
            y='salary_in_usd', 
            data=df, 
            palette='Set2',
            order=['EN', 'MI', 'SE', 'EX'])

plt.title('Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.xticks(ticks=[0,1,2,3], 
           labels=['Entry', 'Mid', 'Senior', 'Executive'])
plt.tight_layout()
plt.show()
```

```python
# Company Size vs Salary
plt.figure(figsize=(10, 6))
sns.boxplot(x='company_size', 
            y='salary_in_usd', 
            data=df, 
            palette='coolwarm',
            order=['S', 'M', 'L'])

plt.title('Salary by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary in USD')
plt.xticks(ticks=[0, 1, 2], 
           labels=['Small', 'Medium', 'Large'])
plt.tight_layout()
plt.show()
```

```python
# Remote Ratio vs Salary
plt.figure(figsize=(10, 6))
sns.boxplot(x='remote_ratio', 
            y='salary_in_usd', 
            data=df, 
            palette='husl')

plt.title('Salary by Remote Work Ratio')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Salary in USD')
plt.xticks(ticks=[0, 1, 2], 
           labels=['On-Site\n(0%)', 'Hybrid\n(50%)', 'Full Remote\n(100%)'])
plt.tight_layout()
plt.show()
```

```python
# Year wise average salary
yearly_salary = df.groupby('work_year')['salary_in_usd'].mean().reset_index()

# Draw the Graph
plt.figure(figsize=(10, 6))
sns.lineplot(x='work_year', 
             y='salary_in_usd', 
             data=yearly_salary,
             marker='o',
             color='steelblue',
             linewidth=2.5)

plt.title('Average Salary Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.tight_layout()
plt.show()
```

```python
# Top 10 highest paying countries
top_countries = df.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(10).reset_index()

# Draw the Graph
plt.figure(figsize=(12, 6))
sns.barplot(x='salary_in_usd', 
            y='company_location', 
            data=top_countries,
            palette='viridis')

plt.title('Top 10 Highest Paying Countries')
plt.xlabel('Average Salary in USD')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
