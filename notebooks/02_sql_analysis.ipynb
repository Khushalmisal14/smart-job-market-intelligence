```python
import pandas as pd
import sqlite3

# Create Database 
conn = sqlite3.connect('../data/processed/job_market.db')

print("✅ Database connected successfully!")
```

```python
# CSV data loading.............
df = pd.read_csv(r'C:\Users\user\Desktop\smart-job-market-intelligence\data\cleaned\ds_salaries_cleaned.csv')

# Save the data into database.........
df.to_sql('jobs', conn, if_exists='replace', index=False)

print("✅ Data loaded into database successfully!")
print(f"Total Rows: {len(df)}")
```

```python
# First SQL Query
query = """
SELECT * FROM jobs LIMIT 5 """

result = pd.read_sql_query(query, conn)
result
```

```python
# Top 10 Highest Paying Job Titles
query = """
SELECT job_title, 
       ROUND(AVG(salary_in_usd), 2) AS avg_salary,
       COUNT(*) AS total_jobs
FROM jobs
GROUP BY job_title
ORDER BY avg_salary DESC
LIMIT 10
"""

result = pd.read_sql_query(query, conn)
result
```

```python
# Experience Level wise Average Salary
query = """
SELECT experience_level,
       ROUND(AVG(salary_in_usd), 2) AS avg_salary,
       COUNT(*) AS total_jobs,
       ROUND(MIN(salary_in_usd), 2) AS min_salary,
       ROUND(MAX(salary_in_usd), 2) AS max_salary
FROM jobs
GROUP BY experience_level
ORDER BY avg_salary DESC
"""

result = pd.read_sql_query(query, conn)
result
```

```python
# Company Size wise Salary Analysis
query = """
SELECT company_size,
       ROUND(AVG(salary_in_usd), 2) AS avg_salary,
       COUNT(*) AS total_jobs
FROM jobs
GROUP BY company_size
ORDER BY avg_salary DESC
"""

result = pd.read_sql_query(query, conn)
result
```

```python
# Remote Work wise Salary Analysis
query = """
SELECT 
    CASE 
        WHEN remote_ratio = 0 THEN 'On-Site'
        WHEN remote_ratio = 50 THEN 'Hybrid'
        WHEN remote_ratio = 100 THEN 'Full Remote'
    END AS work_type,
    ROUND(AVG(salary_in_usd), 2) AS avg_salary,
    COUNT(*) AS total_jobs
FROM jobs
GROUP BY remote_ratio
ORDER BY avg_salary DESC
"""

result = pd.read_sql_query(query, conn)
result
```

```python
# Year wise Salary Trend
query = """
SELECT work_year,
       ROUND(AVG(salary_in_usd), 2) AS avg_salary,
       COUNT(*) AS total_jobs,
       ROUND(MIN(salary_in_usd), 2) AS min_salary,
       ROUND(MAX(salary_in_usd), 2) AS max_salary
FROM jobs
GROUP BY work_year
ORDER BY work_year ASC
"""

result = pd.read_sql_query(query, conn)
result
```
